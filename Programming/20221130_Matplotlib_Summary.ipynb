```python
import matplotlib.pyplot as plt
```

```python
#Line Graph.
plt.plot([0,1,2,3,4], [0,2,4,6,8], label='f(x)=2x', color='red', linewidth=3, linestyle='--', marker='.', markersize=20)
# plt.plot([0,1,2,3], [0,2,4,6], 'b^-')      #color, marker, line

plt.title('Line Graph', fontdict={'fontsize': 20})

plt.xlabel('Number')         #x axis.
plt.ylabel('2 x Number')     #y axis. 

plt.xticks([0,1,2,3,4,5])
plt.yticks([0,2,4,6,8,10])

plt.legend()                 #label required first.

plt.figure(figsize=(5,3))    #resize.

plt.show()
```

```python
#Bar Chart.
plt.bar(['a', 'b', 'c'], [1,4,2])

plt.show()
```

```python
#Histogram.
plt.hist([1,1,1,1,2,2,2,3,3,4,5,8,10], bins=[0,3,6,9])    #bins: set category.
plt.xticks([0,3,6,9])

plt.show()
```

```python
#Pie Chart.
plt.pie([15, 30], labels=['male', 'female'], colors=['red', 'green'], autopct='%.2f %%')

plt.show()
```
